```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_df = pd.read_csv('tmdb/tmdb_5000_movies.csv')

movies_df.sort_values('budget', ascending = False)
```

```python
movies_df.info()
```

```python
# Построим столбчатую диаграмму распределения 20 самых популярных фильмов

movies_barh = movies_df.sort_values('popularity', ascending=False).head(20)
```

```python
plt.barh(movies_barh['original_title'], movies_barh['popularity'])
plt.xlabel('Popularity')
plt.ylabel('Title')
plt.title('Top 20 most popular movies')
```

Построим `boxplot` распределения оценок фильмов (vote_average).

```python
movies_df['vote_average'].plot(kind='box')
```

Интервальный размах распределения оценок находится приблизительное между оценками 5,5 и 6,5 
Максимальные значения распределились между 6,5 и 8,5, минимальные значения - между 3,7 и 5,5.
Среди выбросов видим три случая с очень высокой оценкой, столько же - с очень низкой 
и некоторое количество фильмов с оценкой 2 - 3,8 
На основании данного графика можно предложить критерии 
фильма-"середнячка", фильма-"звезды", "плохого" и "очень плохого" фильма.

Рассмотрим **распределение рейтингов и бюджетов**.

```python
ax = movies_df.vote_average.hist(bins = 20)
ax.set_title('Vote distribution')
ax.set_xlabel('Votes')
ax.set_ylabel('Movies')
```

Распределение оценок приближается к нормальному

```python
ax = movies_df.budget.hist(bins = 50)
ax.set_title('Budget distribution')
ax.set_xlabel('Budget')
ax.set_ylabel('Movies')
```

На этом графике видим значительное превышение количества низкобюджетных фильмов над среднебюджетными,
а также прогрессирующее уменьшение количества фильмов с очень высоким бюджетом.

Построим таблицу с визуализацией дохода в зависимости от производителя(-ей) фильма

```python
top_companies = movies_df.groupby('production_companies')[['revenue']].sum()\
.sort_values('revenue', ascending=False).head(10)
    
top_companies.style.bar()
```



*С помощью seaborn pairplot рассмотрим связь бюджета, дохода и популярности фильмов.*

```python
sns_plot = sns.pairplot(
    movies_df[['budget', 'revenue', 'popularity']])
sns_plot
```

На графике видим достаточно сильную корреляцию бюджета и дохода фильма; 
при этом корреляция популярности и бюджета, популярности и дохода не так очевидна. 

**Распределение оценок** (__`distplot`__)

```python
sns.distplot(movies_df.vote_average)
```

Корреляция оценки и популярности (__`joint_plot`__)

```python
sns.jointplot(x='vote_average', y='popularity', 
              data=movies_df, kind='reg')
# в целом корреляция средней оценки и популярности достаточно заметна
```
